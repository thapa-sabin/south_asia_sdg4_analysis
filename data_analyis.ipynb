```python
df = df.drop('Country Code', axis = 1)
```

```python
df_test = df.groupby(['Series Name', 'Country Name'])[['2008 [YR2008]']].mean()
```

```python
nan_counts = df.isna().sum(axis=1)
nan_counts
```

```python
result = pd.DataFrame({
    "Country Name": df["Country Name"],
    "Series Name": df["Series Name"],
    "NaN Count": nan_counts
})
result.head(8)
```

```python
df.columns[3:]
```

```python
# Determine the columns corresponding to the years
year_columns = df.columns[3:]

# Drop rows with more than 5 NaNs in the year columns
df_filtered = df.dropna(thresh=len(year_columns) - 4, subset=year_columns)

# Print the filtered DataFrame
df_filtered.isna().sum()
```

```python
nan_counts = df_filtered.isna().sum(axis=1)

result = pd.DataFrame({
    "Country Name": df_filtered["Country Name"],
    "Series Name": df_filtered["Series Name"],
    "NaN Count": nan_counts
})
result
```

```python
df_filtered[year_columns] = df_filtered[year_columns].interpolate(axis=1)
df_filtered
```

```python
df_filtered.isna().sum()
```

```python
df_filtered[year_columns] = df_filtered[year_columns].bfill(axis=1)
df_filtered
```

```python
df_filtered.isna().sum()
```

```python
df_filtered["2008 [YR2008]"] = df_filtered[year_columns].mean(axis=1, skipna=True).where(df_filtered["2008 [YR2008]"].isna(), df_filtered["2008 [YR2008]"])
```

```python
df_filtered.isna().sum()
```

```python
df_filtered.reset_index(drop=True)
```
